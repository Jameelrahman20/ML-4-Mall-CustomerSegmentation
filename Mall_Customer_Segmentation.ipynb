

**K-Means Cluster**

Data Visualization & Analysis

```python
X1=customer.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans

#Choosing the number of clusters

#WCSS -> Within Clusters Sum of Squares

wcss=[]

#Elbow Point Plot

for k in range(1,11):
  kmeans=KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(X1)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="8")
plt.title('K Value vs "WCSS')
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)

label=kmeans.fit_predict(X1)

print(label)

c1=kmeans.cluster_centers_

print(c1)
```

```python
#Scatter Plot

plt.scatter(X1[:,0],X1[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(c1[:,0],c1[:,1],color='black')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
X2=customer.loc[:,["Annual Income (k$)","Spending Score (1-100)"]].values

wcss=[]

for k in range(1,11):
  kmeans=KMeans(n_clusters=k, init='k-means++')
  kmeans.fit(X2)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="8")
plt.title('K Value vs "WCSS')
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5)

label=kmeans.fit_predict(X2)

print(label)

c2=kmeans.cluster_centers_

print(c2)
```

```python
plt.scatter(X2[:,0],X2[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(c2[:,0],c2[:,1],color='black')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

Separating features

```python
seg=customer.iloc[:, 2:].values
```

```python
#Apply Standard Scaler

from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()
X_scaled=scaler.fit_transform(seg)
```

Model-1

**K-Means Clustering with Principal Component Analysis(PCA)**

Data Analysis and Evaluation

```python
#Apply PCA

from sklearn.decomposition import PCA

pca=PCA(n_components=2,random_state=42)
X_pca=pca.fit_transform(X_scaled)

#Apply k-means clustering

from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5,random_state=42)
y_kmeans=kmeans.fit_predict(X_scaled)

```

```python
#Calculate the silhouette score

from sklearn.metrics import silhouette_score

s_pca=silhouette_score(X_pca, y_kmeans)*100
print("Silhouette Score of K-Means Clustering with Principal Component Analysis(PCA)  : \n ",s_pca )
```

Model-2

**K-Means Clustering with Singular Value Decomposition(SVD)**

Data Analysis and Evaluation

```python
#Apply SVD

from sklearn.decomposition import TruncatedSVD

svd=TruncatedSVD(n_components=2,random_state=42)
X_svd=svd.fit_transform(X_scaled)

#Apply k-means clustering

from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5,random_state=42)
y_kmeans=kmeans.fit_predict(X_scaled)
```

```python
#Calculate the silhouette score

from sklearn.metrics import silhouette_score

s_svd=silhouette_score(X_svd, y_kmeans)*100
print("Silhouette Score of K-Means Clustering with Singular Value Decomposition(SVD)  : \n ",s_svd )
```

Model-3

**K-Means Clustering**

Data Analysis and Evaluation

```python
#Apply k-means clustering

from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5,random_state=42)
y_kmeans=kmeans.fit_predict(X_scaled)
```

```python
#Calculate the silhouette score

from sklearn.metrics import silhouette_score

s_kmc=silhouette_score(X_scaled, y_kmeans)*100

print("Silhouette Score for K-Means Clustering : \n ", s_kmc)
```
